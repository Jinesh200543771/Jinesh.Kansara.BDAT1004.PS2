```python
class Marsupial:                    #Super class method which will append an item to the empty list items[]
   def __init__(self):
       self.items = []


   def put_in_pouch(self, item):    #function put_in_pouch will append items to the list
       self.items.append(item)


   def pouch_contents(self):        #function pouch_contents will print the appended list
       print(self.items)




m = Marsupial()
m.put_in_pouch("doll")
m.put_in_pouch("firestruck")
m.put_in_pouch("kitten")
m.pouch_contents()
```

```python
def collatz(x):
    if x == 1:               #if x=1, function will stop running and return 1
        return [x]
    
    elif x%2==0:             #Check weather the given input is even or not
        return [x] + collatz(int(x/2))
    
    else:                    #check weather the given input is odd or not
        return [x] + collatz(int(x*3+1))


resultA=collatz(1)
resultB=collatz(10)
print("collatz(1)")
print(*resultA,sep = "\n") #Sep parameter to valuein new line
print("\ncollatz(10)")
print(*resultB, sep = "\n")
```

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    Headings = ["h1","h2"]
    current = -1
    
    def handle_starttag(self, tag, attrs):
        if tag in HeadingParser.Headings:     #check for headers from the Heading List
            HeadingParser.current = HeadingParser.Headings.index(tag)
    
    def handle_endtag(self,tag):               # End tag Handler
        if tag == HeadingParser.Headings[HeadingParser.current]:
            HeadingParser.Current = -1
            
    def handle_data(self,data):    #setting Printing identation for Heading tags
        if HeadingParser.current >= 0:
            print(" "*HeadingParser.current+data)
            
```

```python
from urllib.request import urlopen
from urllib.parse import urljoin
from html.parser import HTMLParser

class Collector(HTMLParser):
   # collects hyperlink URLs into a list


   def __init__(self, url):
       # initializes parser, the url, and a list
       HTMLParser.__init__(self)
       self.url = url
       self.links = []


   def handle_starttag(self, tag, attrs):
       # collects hyperlink URLs in their absolute format    (Reference Class Notes Week 6 Example)
       if tag == 'a':  # checking for anchor tag
           for attr in attrs:
               if attr[0] == 'href':
                   # construct absolute URL
                   absolute = urljoin(self.url, attr[1])
                   if absolute[:4] == 'http':  # collect HTTP URLs
                       self.links.append(absolute)  # appending all the links in a list


   # Recursive webdir function according depth and indent


   def webdir(self, url, depth, indent):
       print(indent * " " + url)
       if depth == 0:
           return
       for i in self.links:
           self.webdir(i, depth - 1, indent + 1)
```
